```python
import pandas as pd

# Load raw dataset
df = pd.read_csv("superstore.csv")  

df.head()
```

```python
df.columns
```

```python
customers = df[
    ['Customer ID', 'Customer Name', 'Segment']
].drop_duplicates()

customers.columns = [
    'customer_id',
    'customer_name',
    'segment'
]

customers.to_csv('customers.csv', index=False)
```

```python
products = df[
    ['Product ID', 'Product Name', 'Category', 'Sub-Category']
].drop_duplicates()

products.columns = [
    'product_id',
    'product_name',
    'category',
    'sub_category'
]

products.to_csv('products.csv', index=False)
```

```python
regions = df[
    ['Country', 'Region', 'State', 'City', 'Postal Code']
].drop_duplicates()

regions['region_id'] = range(1, len(regions) + 1)

regions = regions[
    ['region_id', 'Country', 'Region', 'State', 'City', 'Postal Code']
]

regions.columns = [
    'region_id',
    'country',
    'region',
    'state',
    'city',
    'postal_code'
]

regions.to_csv('regions.csv', index=False)
```

```python
orders = df[
    [
        'Order ID',
        'Order Date',
        'Ship Date',
        'Ship Mode',
        'Customer ID',
        'Product ID',
        'Country',
        'Region',
        'State',
        'City',
        'Postal Code',
        'Sales'
    ]
]

orders.columns = [
    'order_id',
    'order_date',
    'ship_date',
    'ship_mode',
    'customer_id',
    'product_id',
    'country',
    'region',
    'state',
    'city',
    'postal_code',
    'sales'
]

# Map region_id
orders = orders.merge(
    regions,
    on=['country', 'region', 'state', 'city', 'postal_code'],
    how='left'
)

orders = orders.drop(
    columns=['country', 'region', 'state', 'city', 'postal_code']
)

orders.to_csv('orders.csv', index=False)
```
